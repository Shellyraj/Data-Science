```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/kaggle/input/datasets/mohammadtalib786/retail-sales-dataset/retail_sales_dataset.csv")

# Group by Product Category
grouped = df.groupby('Product Category')['Total Amount'].sum()

# Print results
print(grouped)

# Bar Graph
plt.figure()
grouped.plot(kind='bar')
plt.title('Purchase Behavior by Category')
plt.xlabel('Product Category')
plt.ylabel('Total Purchase Amount')
plt.show()

# Pie Chart
plt.figure()
grouped.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Purchase Distribution by Category')
plt.show()
```
